```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import yeyo as yy
```

```python
#a) Carga de archivos (csv, xlsx, html y json) a dataframes .
df = yy.cargar_dataset('Porto_Portugal.csv')
yy.cuenta_valores_nulos(df)
```

```python
#b) Sustitución de valores nulos con el método de “ffill”.
ffill = yy.sust_ffill(df)
yy.cuenta_valores_nulos(ffill)
```

```python
#c) Sustitución de valores nulos con el método de “bfill”.
bfill = yy.sust_bfill(df)
yy.cuenta_valores_nulos(bfill)
```

```python
#d) Sustitución de valores nulos con el método por un “string concreto”
string = yy.sust_string(df,'within a day')
yy.cuenta_valores_nulos(string)
```

```python
#e) Sustitución de valores nulos con el método de “promedio”.
promedio = yy.sust_prom(df)
yy.cuenta_valores_nulos(promedio)
```

```python
#f) Sustitución de valores nulos con el método de “mediana”
mediana = yy.sust_median(df)
yy.cuenta_valores_nulos(mediana)
```

```python
#g) Sustitución de valores nulos con el método de “constante”.
constante = yy.sust_constante(df,10)
yy.cuenta_valores_nulos(constante)
```
